```python
import requests
import xmltodict
from datetime import datetime
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import pprint
import ast
import json
from io import StringIO
pp = pprint.PrettyPrinter(indent=2)
```

```python
passing_base_url = "https://premium.pff.com/api/v1/facet/passing/summary?league=nfl&season=2020&export=true"
passing_df = pd.DataFrame()
for i in range(1, 18):
    url = f"{passing_base_url}&week={i}"
    r = requests.get(url)
    print(r)
    week_stats = pd.read_csv(StringIO(r.text)).dropna(axis=1, how="all")
    week_stats["week"] = i
    passing_df = passing_df.append(week_stats, ignore_index=True)
```

```python
passing_df = passing_df.drop(columns="player_game_count").rename(columns={"attempts": "passing_attempts",
                                                             "yards": "passing_yards",
                                                             "touchdowns": "passing_tds",
                                                             "ypa": "passing_ypa"})
```

```python
receiving_base_url = "https://premium.pff.com/api/v1/facet/receiving/summary?league=nfl&season=2020&export=true"
receiving_df = pd.DataFrame()
for i in range(1, 18):
    url = f"{receiving_base_url}&week={i}"
    r = requests.get(url)
    print(r)
    week_stats = pd.read_csv(StringIO(r.text)).dropna(axis=1, how="all")
    week_stats["week"] = i
    receiving_df = receiving_df.append(week_stats, ignore_index=True)
```

```python
receiving_df = receiving_df.drop(columns="player_game_count").rename(columns={"yards": "receiving_yards",
                                                                              "touchdowns": "receiving_touchdowns"})
```

```python
rushing_base_url = "https://premium.pff.com/api/v1/facet/rushing/summary?league=nfl&season=2020&export=true"
rushing_df = pd.DataFrame()
for i in range(1, 18):
    url = f"{rushing_base_url}&week={i}"
    r = requests.get(url)
    print(r)
    week_stats = pd.read_csv(StringIO(r.text)).dropna(axis=1, how="all")
    week_stats["week"] = i
    rushing_df = rushing_df.append(week_stats, ignore_index=True)
```

```python
rushing_df
```

```python
rushing_df = rushing_df.drop(columns="player_game_count").rename(columns={"attempts": "rushing_attempts",
                                                             "yards": "rushing_yards",
                                                             "touchdowns": "rushing_touchdowns",
                                                             "ypa": "rushing_ypa",
                                                             "longest": "rushing_longest"})
```

```python
kicking_base_url = "https://www.pff.com/api/fantasy/stats/kicking?season=2020"

kicking_df = pd.DataFrame()
for i in range(1, 18):
    url = f"{kicking_base_url}&weeks={i}"
    r = requests.get(url)
    print(r)
    week_stats = pd.read_json(StringIO(r.text))
    week_stats["week"] = i
    kicking_df = kicking_df.append(week_stats, ignore_index=True)
```

```python
kicking_df["position"] = "K"
kicking_df = kicking_df.rename(columns={"team": "team_name"})
```

```python
dst_base_url = "https://www.pff.com/api/fantasy/stats/dst?season=2020"

dst_df = pd.DataFrame()
for i in range(1, 18):
    url = f"{dst_base_url}&weeks={i}"
    r = requests.get(url)
    print(r)
    week_stats = pd.read_json(StringIO(r.text))
    week_stats["week"] = i
    dst_df = dst_df.append(week_stats, ignore_index=True)
```

```python
dst_df["position"] = "DST"
dst_df = dst_df.rename(columns={"team": "player"})
dst_df["team_name"] = dst_df["player"]
```

```python
merged_df = passing_df.merge(rushing_df,
                             how="outer",
                             on=["player", "player_id", "position","team_name", "week"])
merged_df = merged_df.merge(receiving_df,
                             how="outer",
                             on=["player", "player_id", "position","team_name", "week"])
merged_df = merged_df.merge(kicking_df,
                             how="outer",
                             on=["player", "player_id", "position","team_name", "week"])
merged_df = merged_df.merge(kicking_df,
                             how="outer",
                             on=["player", "player_id", "position","team_name", "week"])
merged_df = merged_df.merge(dst_df,
                             how="outer",
                             on=["player", "position","team_name", "week"])
```

```python
rushing_df.loc[(rushing_df["position"] == "HB") | (rushing_df["position"] == "FB"), "position"] = "RB"
```
