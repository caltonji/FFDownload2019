```python
positions = ['QB', 'QB', 'WR', 'WR','WR', 'RB', 'RB', 'TE', 'WR/RB/TE', 'K', 'DEF']
def optimal_points(players, team_name, week):
    week_players = players.loc[(players['team_name'] == team_name) & (players['week'] == week)]
    points = 0.0
    for position in positions:
        allowed_positions = position.split("/")
        players_at_position = week_players.loc[players["position"].isin(allowed_positions)]
        best_player = players_at_position.iloc[players_at_position["points"].argmax()]
        week_players = week_players.loc[week_players["player_key"] != best_player["player_key"]]
        points += best_player["points"]
    return points
```

```python
positions = ['QB', 'QB', 'WR', 'WR','WR', 'RB', 'RB', 'TE', 'WR/RB/TE', 'K', 'DEF']
def optimal_player_points(players):
    points = 0.0
    player_dict = players.reset_index().to_dict()
    for position in positions:
        maxPoints = -1.0
        maxIndex = -1
        allowed_positions = position.split("/")
        for i in player_dict["position"].keys():
            if player_dict["position"][i] in allowed_positions and player_dict["points"][i] > maxPoints:
                maxPoints = player_dict["points"][i]
                maxIndex = i
        if (maxIndex > 0):
            points += maxPoints
            del player_dict["position"][maxIndex]
            del player_dict["points"][maxIndex]
    return points
```

```python
def actual_points(players, team_name, week):
    return players.loc[(players['started']) & (players['team_name'] == team_name) & (players['week'] == week)]['points'].sum()
```

```python
import time
```

```python
start = time.time()

weekPointsDicts = []
for index, team in teams.iterrows():
    teamName = team["name"]
    for week in range(1, 17):
        actual = actual_points(players, teamName, week)
        optimal = optimal_points(players, teamName, week)
        weekPointsDict = {
            "team_name": teamName,
            "actual_points": actual,
            "optimal_points": optimal
        }
        weekPointsDicts.append(weekPointsDict)
weekPointsDf = pd.DataFrame(weekPointsDicts,
                     columns=['team_name', 'actual_points', 'optimal_points'])
end = time.time()
print(end - start)
weekPointsDf
```

```python
start = time.time()
groupByTeamByWeek = players.groupby(['team_name', 'week'])
weeks = pd.DataFrame({'actual' : groupByTeamByWeek.apply(lambda x: x[x["started"]]["points"].sum()),
                      'optimal' : groupByTeamByWeek.apply(optimal_player_points)}).reset_index()
end = time.time()
print(f"Time {end - start}")
```
